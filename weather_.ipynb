```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Weather_Data.csv")
```

```python
df = df.dropna()
```

```python
print(df.head())
```

```python
hot = df[df["MaxTemp"] > 30]

print("\nHot Weather:\n")
print(hot)
```

```python
temp = np.array(df["MaxTemp"])

print("\nAverage Temp:", np.mean(temp))
print("Max Temp:", np.max(temp))
print("Min Temp:", np.min(temp))
```

```python
group = df.groupby("RainToday")["MaxTemp"].mean()

print("\nAverage Max Temperature by RainToday:\n")
print(group)
```

```python
group.plot(kind="bar")

plt.title("Weather vs Temperature")
plt.xlabel("Weather")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.plot(df["MaxTemp"])

plt.title("Max Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Max Temperature")
plt.show()
```

```python
df["RainToday"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("RainToday Distribution")
plt.ylabel("")
plt.show()
```
