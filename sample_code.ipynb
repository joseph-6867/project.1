```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Employee.csv')
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
df = df.dropna()
```

```python
column = input("\nEnter column to filter (): ")
```

```python
print("\nChoose condition:")
print("1 - Greater than")
print("2 - Less than")
print("3 - Equal to")
```

```python
choice = input("Enter choice: ")
value = input("Enter value: ")
```

```python
if value.replace('.', '', 1).isdigit():
    value = float(value)
```

```python
if column not in df.columns:
    found_col = None
    for col in df.columns:
        if col.lower() == column.lower():
            found_col = col
            break
    if found_col:
        column = found_col
    else:
        # If no column found, let the KeyError propagate or handle more gracefully
        # For now, we'll assume a case-insensitive match is desired if exact match fails.
        pass

if choice == "1":
    result = df[df[column] > value]
elif choice == "2":
    result = df[df[column] < value]
else:
    result = df[df[column] == value]
```

```python
print(result)
```

```python
if "Salary" in df.columns:
    sorted_data = result.sort_values(by="Salary", ascending=False)
    print("\nSorted by Salary (High to Low):\n", sorted_data)
```
