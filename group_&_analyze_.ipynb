```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Employee.csv")
```

```python
print("Average Age:",df["Age"].mean())
```

```python
df = df.dropna()
```

```python
city_group = df.groupby("City")

```

```python
city_analysis = city_group.agg({
    "Age": ["mean", "max", "min"],
    "ExperienceInCurrentDomain": ["mean", "max"],
    "PaymentTier": "count"
})
```

```python
print(city_analysis)
```

```python
tier_group = df.groupby("PaymentTier")
```

```python
tier_analysis = tier_group.agg({
    "Age": ["mean", "max", "min"],
    "ExperienceInCurrentDomain": ["mean", "max"],
    "City": "count"
})
```

```python
print(tier_analysis)
```

```python
gender_group = df.groupby("Gender")
```

```python
gender_group = df.groupby("Gender")

gender_analysis = gender_group.agg({
    "Age": ["mean", "max"],
    "ExperienceInCurrentDomain": "mean",
    "City": "count"
})
```

```python
print(gender_analysis)
```

```python
multi_analysis = df.groupby(["City", "Gender"]).agg({
    "Age": "mean",
    "ExperienceInCurrentDomain": "mean"
})
```

```python
print(multi_analysis)
```
