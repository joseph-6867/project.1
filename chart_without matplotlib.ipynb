### Step 1: Save your Streamlit code to a `.py` file

First, we need to save the Streamlit code into a Python file. This is because `streamlit run` command expects a file as input.

```python
%%writefile streamlit_app.py

import streamlit as st
import pandas as pd
import plotly.express as px

# Sample data
data = {
    "Days": [1, 2, 3, 4, 5],
    "Sales": [100, 200, 150, 300, 250]
}

df = pd.DataFrame(data)

st.title("Sales Line Chart")

# Create line chart
fig_line = px.line(df, x="Days", y="Sales", title="Daily Sales (Line Chart)")
st.plotly_chart(fig_line)

st.title("Sales Bar Chart")

# Create bar chart
fig_bar = px.bar(df, x="Days", y="Sales", title="Daily Sales (Bar Chart)")
st.plotly_chart(fig_bar)

st.title("Sales Scatter Chart")

# Create scatter chart
fig_scatter = px.scatter(df, x="Days", y="Sales", title="Daily Sales (Scatter Chart)")
st.plotly_chart(fig_scatter)
```
